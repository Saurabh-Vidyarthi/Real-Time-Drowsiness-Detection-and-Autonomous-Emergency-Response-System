```python
import torch
from torchvision.models import resnet18
from torch.nn import Linear, Sequential

# Define the file path where the model is saved
model_path = "model_bf16.pth"

# Define the model architecture
model_loaded = Sequential(
    resnet18(pretrained=True),
    Linear(1000, 2)  # Assuming 2 output classes
)

# Load the model state dictionary
model_state_dict = torch.load(model_path)

# Convert all parameters to bfloat16
model_bf16 = model_loaded.to(torch.bfloat16)

# Load the modified state dictionary into the model
model_loaded.load_state_dict(model_state_dict)

print(f"Model loaded from {model_path}")
```

```python
# Iterate over the parameters and print their data types
for name, param in model_loaded.named_parameters():
    print(f"Parameter '{name}' has data type: {param.dtype}")
```

```python
# Convert all parameters to bfloat16
for param_name in model_state_dict.keys():
    model_state_dict[param_name] = model_state_dict[param_name].to(torch.bfloat16)
    
# Iterate over the parameters and print their data types
for name, param in model_loaded.named_parameters():
    print(f"Parameter '{name}' has data type: {param.dtype}")
```

```python
def print_param_dtype(model):
    for name, param in model.named_parameters():
        print(f"{name} is loaded in {param.dtype}")
```
